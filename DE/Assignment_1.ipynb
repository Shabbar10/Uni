### To check if there are any duplicated values in the DataFrame:

```python
df.duplicated().values.any()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().values.any()
```

### To see the number of rows in each column that contain a missing value

```python
df.isnull().sum()
```

### Here we try to replace the missing value with the mode of the column. While this works for numerical columns, it does no for string columns.

```python
df = df.fillna(df.mode())
```

```python
df.isnull().sum()
```

### As such we replace the remaining missing values with a constant i.e. 0.

```python
df = df.fillna(0)
```

```python
df.isnull().sum()
```
