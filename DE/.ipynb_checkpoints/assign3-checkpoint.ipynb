```python
year_data = df["published_year"].tolist()

year_data_final = []

for x in year_data:
    if isinstance(x, float) and np.isnan(x):
        year_data_final.append(x)  # Keep NaN as is
    else:
        year_data_final.append(int(x))  # Convert other values to int

year_data_final = pd.Series(year_data_final, dtype=pd.Int64Dtype())


df["published_year"] = year_data_final

```

```python
def myBin(arr, nbins : int):
    bin_array = []

    if not isinstance(arr, pd.Series):
        raise TypeError("Argument must be a pandas series")

    try:
        if isinstance(arr, pd.Series):
            bin_array = arr.values

        # Checking for possible missing values in the input data
        nan_values = bin_array.isna().sum()

        # Since, pandas categorizes nan values in different bin for specifically NaN values, we would try to do the same.
        nan_bin = np.isnan(bin_array)
        bin_array = bin_array[~nan_bin]

        # Equal-width binning
        width = (max(bin_array) - min(bin_array)) / nbins


        # [min + w], [min + 2w], [min + 3w] .... 
        # Forming bins of equal width
        bins = []

        for i in range(nbins):
            min_value = bin_array.min()
            lower_bound = min_value + (i * width)
            upper_bound = lower_bound + width

            # Bins using Interval variable
            inv = pd.Interval(left=lower_bound, right=upper_bound, closed="left")

            bins.append([value for value in bin_array if value in inv])

        return bins

    except Exception as e:
        print(f"Error : {e}")
```

```python
bin1 = myBin(df["published_year"], nbins=4)

for inner_bin in bin1:
    print("Values in current list: ")

    print(f"[{min(inner_bin)}, {max(inner_bin)}]")

    print()
```
